```python
# Please install OpenAI SDK first: `pip3 install openai`
from dotenv import load_dotenv
import os

load_dotenv()

API_KEY = os.getenv("DEEPSEEK_API_KEY")
from openai import OpenAI

client = OpenAI(api_key=API_KEY, base_url="https://api.deepseek.com")

prompt = f"""
        User is feeling Sad.
        User is looking for a way to improve mood.

        Here are the locally installed apps and their paths that can help:
        available_apps - whatsapp, skype, teams, discord, zoom, spotify, youtube

        There are two outputs. 
        - 'recommendation': Suggestion to improve the mood. Give the most suitable 3 recommandations each containing 4 words, according to the selected apps and online available apps(like youtube, spotify, online games like free sites). 
        - 'recommendation_options': list of 2 apps that are available locally or online.
        The recommendation_options should be apps. It contains 3 parameters:
        - app_name: Name of the app
        - app_url: URL of the app ('https://xxxxxxx.com') or path of the app from above available_apps
        - search_query: If the app is a web browser, give a suitable search query to find the app.
        Response Formate:
        '''[
            {{
                "recommendation": "",
                "recommendation_options": [{{
                    "app_name": "",
                    "app_url": "",
                    "search_query": ""
                }}],
                
            }}]'''
        Respond ONLY with the exact phrase from the list.
"""


response = client.chat.completions.create(
    model="deepseek-reasoner",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": prompt},
    ],
    stream=False
)

print(response.choices[0].message.content)
```

```python
import webbrowser
import psutil
import win32gui
import win32con
import win32process
import time

def is_browser_running(browser_name):
    for proc in psutil.process_iter(['pid', 'name']):
        if proc.info['name'] and browser_name.lower() in proc.info['name'].lower():
            return proc.info['pid']
    return None

def bring_browser_to_front(pid):
    def enumHandler(hwnd, lParam):
        if win32gui.IsWindowVisible(hwnd):
            _, found_pid = win32process.GetWindowThreadProcessId(hwnd)
            if found_pid == pid:
                win32gui.ShowWindow(hwnd, win32con.SW_RESTORE)
                win32gui.SetForegroundWindow(hwnd)
                return False
        return True
    win32gui.EnumWindows(enumHandler, None)

def open_url_in_browser(url, browser_path=None):
    if browser_path:
        webbrowser.register('custom', None, webbrowser.BackgroundBrowser(browser_path))
        browser = webbrowser.get('custom')
    else:
        browser = webbrowser
    browser.open(url, new=2)

url = "https://youtube.com"
browser_exe = "chrome.exe"
chrome_path = r"C:\Program Files\Google\Chrome\Application\chrome.exe"

pid = is_browser_running(browser_exe)
open_url_in_browser(url, browser_path=chrome_path)

# Sleep to allow browser to open new tab
time.sleep(2)

if pid:
    bring_browser_to_front(pid)
else:
    print("Browser was not running previously.")
```

```python
import sys, subprocess, time, urllib.request, json
import psutil, win32gui, win32con, win32process
from websocket import create_connection
import webbrowser, random, re

def start_chrome_debug(chrome_exe, data_dir=None):
    # Launch an entirely new Chrome profile with remote debugging.
    port = random.choice([9222, 9333, 9444])
    cmd = [chrome_exe, f"--remote-debugging-port={port}"]
    if data_dir:
        cmd.append(f"--user-data-dir={data_dir}")
    subprocess.Popen(cmd)
    time.sleep(3)
    return port

def open_new_tab_via_cdp(port, url):
    info = urllib.request.urlopen(f"http://localhost:{port}/json/new?{url}", timeout=3)
    tab = json.loads(info.read().decode())
    return tab["id"], tab["webSocketDebuggerUrl"]

def close_target_via_cdp(ws_url, target_id):
    ws = create_connection(ws_url)
    ws.send(json.dumps({
        "id": 1,
        "method": "Target.closeTarget",
        "params": {"targetId": target_id}
    }))
    ws.close()
    return True

def is_chrome_running(exe_name="chrome.exe"):
    for p in psutil.process_iter(['pid', 'name']):
        if p.info['name'] and exe_name.lower() in p.info['name'].lower():
            return p.info['pid']
    return None

def bring_to_front(pid):
    def handler(hwnd, _):
        if win32gui.IsWindowVisible(hwnd):
            _, found = win32process.GetWindowThreadProcessId(hwnd)
            if found == pid:
                win32gui.ShowWindow(hwnd, win32con.SW_RESTORE)
                win32gui.SetForegroundWindow(hwnd)
                return False
        return True
    win32gui.EnumWindows(handler, None)

def main():
    chrome_exe = r"C:\Program Files\Google\Chrome\Application\chrome.exe"
    custom_profile = r"C:\temp\cdp-profile"
    url = "https://youtube.com"
    
    pid = is_chrome_running()
    debug_port = None
    
    # Try existing debug or launch new Chrome
    for p in [9222, 9333, 9444]:
        try:
            tabs = urllib.request.urlopen(f"http://localhost:{p}/json", timeout=1)
            debug_port = p
            break
        except:
            continue
    if not debug_port:
        debug_port = start_chrome_debug(chrome_exe, data_dir=custom_profile)
    
    print(f"[+] Using CDP port {debug_port}")
    
    target_id, ws_url = open_new_tab_via_cdp(debug_port, url)
    print(f"[+] YouTube opened in tab ID: {target_id}")
    
    print("[*] Waiting 5 seconds before closing YouTube tab...")
    time.sleep(5)
    
    if close_target_via_cdp(ws_url, target_id):
        print("[] YouTube tab closed cleanly.")
    else:
        print("[!] Failed to close the tab via CDP.")
    
    if pid:
        time.sleep(1)
        bring_to_front(pid)

if __name__ == "__main__":
    main()
```
